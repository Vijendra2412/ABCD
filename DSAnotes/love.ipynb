Array ek basic data structure hai jisme multiple elements ko ek saath store kiya ja sakta hai. Informally, array ek list hoti hai jisme same type ke elements store hote hain, jaise ki numbers, words, ya even doosre arrays. Har element ka ek index hota hai, jo 0 se start hota hai. Example ke liye, agar humare paas array [1, 2, 3] hai, toh 1 ka index 0, 2 ka index 1, aur 3 ka index 2 hoga.

Memory mein Array ka Structure:-

Array memory mein contiguous blocks mein store hoti hai, iska matlab hai sabhi elements ek ke baad ek store hote hain. Jab hum kisi index ko access karte hain, toh memory ke base address se offset calculate karke value ko retrieve kiya jata hai. Ye kaafi efficient hota hai.

2D Arrays:-


2D arrays ek table ya matrix jaise structure ko represent karte hain jisme rows aur columns hote hain. Ye kaafi useful hote hain jab hume grid-based problems solve karni hoti hai, jaise graph traversal ya dynamic programming ke questions.



Yahaan pe matrix[1][2] ka matlab hai second row ka teesra element, jo ki 6 hai.
