Hash Tables aur Hash Maps

1. Hash Table ki Definition
Hash Table: Ek data structure jo key-value pairs ko store karne ke liye use hota hai. Ye keys ko values se map karta hai, jisse data ko efficiently store aur retrieve kiya ja sakta hai.
2. Key Characteristics
Time Complexity:
Average-case mein insert, delete, aur search operations O(1) time mein hoti hain.

Collision Handling:
Chaining: Jab do keys same index par hash hoti hain, tab har bucket ek linked list ko hold karta hai jisme saari entries hoti hain.

Open Addressing: Collision hone par, agle available slot ko dhoondha jata hai.

3. Java mein HashMap
HashMap ek implementation hai Map interface ka, jo Java Collections Framework ka part hai.

Key Features:
Non-synchronized: Ye thread-safe nahi hai, isliye multiple threads is par kaam karte waqt dikkat aa sakti hai.

Null Values: Ek null key aur multiple null values ko allow karta hai.

Unordered: Ye elements ko kisi particular order mein nahi store karta.

4. HashMap ka Design

Core Methods:

put(key, value): Key-value pair ko insert ya update karta hai.

get(key): Diye gaye key ka value retrieve karta hai.

remove(key): Specified key ka key-value pair delete karta hai.

Approach:

Explanation: HashMap mein har element ka index store karte hain. Jab hum har element ka complement check karte hain (target - current element), agar wo map mein milta hai toh voh indices return karte hain.

Remove Duplicates from Sorted Array

Problem Statement

Aapko ek sorted array diya gaya hai aur aapko us array se duplicate elements ko remove karna hai. Aapko yeh operation in-place karna hai (matlab bina extra space use kiye) aur final count of unique elements return karna hai.

Approach
Brute Force Approach:

Naya array banayein aur original array ko traverse karein. Jo elements value ke alawa hain, unhe naye array mein daal dein.

Lekin, humein in-place solution chahiye, isliye humein is approach ko avoid karna hoga.

In-Place Approach:

Do pointers ka istemal karte hue, ek pointer (slow) unique elements ka track rakhega aur dusra pointer (fast) array ko traverse karega.

Agar fast pointer ka element slow pointer ke element se alag hai, to slow pointer ko increment karein aur nums[slow] mein nums[fast] ki value assign karein.

Yeh process tab tak chalta rahega jab tak fast pointer array ke end tak nahi pahunch jaata.

Line 9: Fast pointer se array ko traverse kiya jata hai.
